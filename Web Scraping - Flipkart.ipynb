```python
import bs4
import pandas as pd
from urllib.request import urlopen as uReq
from bs4 import BeautifulSoup as soup
```

```python
myurl='https://www.flipkart.com/search?q=iPhone11&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off'
myurl
```

```python
uclient = uReq(myurl)
uclient_html = uclient.read()
uclient.close()
page_soup = soup(uclient_html,"html.parser")
```

```python
containers = page_soup.findAll("div",{"class":"_1-2Iqu row"})
container = containers[0]
f_name = "arima.csv"
f = open(f_name,'w')
headers = "Phone_Details,Phone_price\n"
f.write(headers)
for container in containers:
    container1 = container.findAll("div",{"class":"_3wU53n"})
    container2 = container.findAll("div",{"class":"_1vC4OE _2rQ-NK"})
    phone_detail = container1[0].getText()
    phone_price  = container2[0].getText()
    phone_detail = phone_detail.replace(',',' ')
    phone_price = phone_price.replace(',','')
    phone_price = phone_price.replace('âÃÏ','Rs')
    f.write(phone_detail + "," + phone_price + "\n")
f.close()
```
